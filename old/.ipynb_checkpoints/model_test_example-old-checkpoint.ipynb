```python
import logging
import os

import torch
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

import pyro
import graphviz
import pyro.distributions as dist
import pyro.distributions.constraints as constraints
```

```python
smoke_test = ('CI' in os.environ)
assert pyro.__version__.startswith('1.8.4')

pyro.enable_validation(True)
pyro.set_rng_seed(1)
logging.basicConfig(format='%(message)s', level=logging.INFO)

# Set matplotlib settings
%matplotlib inline
plt.style.use('default')
```

```python
# Generate data
s = np.array([[1.0, 2.0]])
d = np.array([[3.0, 4.0]])
```

```python
# Compute observed data
obs_data = torch.tensor(np.matmul(np.transpose(s), d))
mask = torch.tensor([[True, True], [True, False]])
print(s)
print(d)
print("obs_data: ")
print(obs_data)
print("mask: ")
print(mask)
```

```python
# Add noise
noise = torch.tensor(np.random.normal(0, 1, size=(2, 2)))
obs_data += noise
print(obs_data)
```

```python
def model(obs_data, mask):
    s1 = pyro.sample("s1", dist.Normal(1, 1))
    s2 = pyro.sample("s2", dist.Normal(2, 1))
    d1 = pyro.sample("d1", dist.Normal(3, 1))
    d2 = pyro.sample("d2", dist.Normal(4, 10))
    s = [s1, s2]
    d = [d1, d2]
    nsamples, ndrugs = obs_data.shape
    result = torch.empty((nsamples, ndrugs), dtype=torch.int32)
    sigma = pyro.param("sigma", lambda: torch.ones(()), constraint=constraints.positive)
    for i in range(nsamples):
        for j in range(ndrugs):
            result[i, j] = pyro.sample(f"obs_{i,j}", 
                                       dist.Normal(s[i] * d[j], sigma), 
                                       obs=obs_data[i,j] if mask[i,j] else None)
    return result
```

```python
pyro.render_model(model, model_args=(obs_data, mask), render_distributions=True)
```

```python
pyro.clear_param_store()
kernel = pyro.infer.mcmc.NUTS(model, jit_compile=True)
mcmc = pyro.infer.MCMC(kernel, num_samples=500, warmup_steps=500)
mcmc.run(obs_data, mask)
```

```python
mcmc_samples = {k: v.detach().cpu().numpy() for k, v in mcmc.get_samples().items()}
```

```python
fig, axes = plt.subplots(1, 5, figsize=(15, 5))

sns.histplot(mcmc_samples['s1'], ax=axes[0])
axes[0].set_xlabel('s1')

sns.histplot(mcmc_samples['s2'].flatten(), ax=axes[1])
axes[1].set_xlabel('s2')

sns.histplot(mcmc_samples['d1'].flatten(), ax=axes[2])
axes[2].set_xlabel('d1')

sns.histplot(mcmc_samples['d2'].flatten(), ax=axes[3])
axes[3].set_xlabel('d2')

sns.histplot(mcmc_samples['obs_(1, 1)'].flatten(), ax=axes[4])
axes[4].set_xlabel('obs_(1, 1)')
```

```python
d1 = torch.tensor([1])

def model():
    a = pyro.sample("a", dist.Normal(0, 1))
    b = pyro.sample("b", dist.Normal(0, 1))
    sigma = pyro.param("sigma", lambda: torch.ones(()), constraint=constraints.positive)
    mean = a + b
    return pyro.sample("obs", dist.Normal(mean, sigma), obs=d1)

pyro.render_model(model, render_distributions=True)
```

```python
d = torch.tensor([1, 2])

def model(d):
    a = pyro.sample("a", dist.Normal(0, 1))
    b = pyro.sample("b", dist.Normal(0, 1))
    c = pyro.sample("c", dist.Normal(0, 1))
    sigma = pyro.param("sigma", lambda: torch.ones(()), constraint=constraints.positive)
    mean1 = a * b
    mean2 = b * c
    means = [mean1, mean2]
    result = torch.empty_like(d)
    for i in range(len(d)):
        result[i] = pyro.sample(f"obs_{i}", dist.Normal(means[i], sigma), obs=d[i] if d is not None else None)
    return result

pyro.render_model(model(d), render_distributions=True)
```

```python
data_table = torch.tensor([[1, 2], [3, 4], [5, 6]])

def model():
    s0 = pyro.sample("s0", dist.Normal(0, 1))
    s1 = pyro.sample("s1", dist.Normal(0, 1))
    s2 = pyro.sample("s2", dist.Normal(0, 1))
    samples = [s0, s1, s2]
    d0 = pyro.sample("d0", dist.Normal(0, 1))
    d1 = pyro.sample("d1", dist.Normal(0, 1))
    drugs = [d0, d1]
    sigma = pyro.param("sigma", lambda: torch.ones(()), constraint=constraints.positive)
    
    result = torch.empty_like(data_table)
    result[0, 0] = pyro.sample(f"obs_{0,0}", 
                               dist.Normal(s0 * d0, sigma), 
                               obs=data_table[0][0] if data_table is not None else None)
    result[0, 1] = pyro.sample(f"obs_{0,1}", 
                               dist.Normal(s0 * d1, sigma), 
                               obs=data_table[0][1] if data_table is not None else None)
    result[1, 0] = pyro.sample(f"obs_{1,0}", 
                               dist.Normal(s1 * d0, sigma), 
                               obs=data_table[1][0] if data_table is not None else None)
    result[1, 1] = pyro.sample(f"obs_{1,1}", 
                               dist.Normal(s1 * d1, sigma), 
                               obs=data_table[1][1] if data_table is not None else None)
    result[2, 0] = pyro.sample(f"obs_{2,0}", 
                               dist.Normal(s2 * d0, sigma), 
                               obs=data_table[2][0] if data_table is not None else None)
    result[2, 1] = pyro.sample(f"obs_{2,1}", 
                               dist.Normal(s2 * d1, sigma), 
                               obs=data_table[2][1] if data_table is not None else None)
    return result

pyro.render_model(model, render_distributions=True)
```
