```python
# in run_2c data....
# choose pairs to highlight: for each target, select the source w the highest raw_overlap baseline
# groupby target, source, compute mean of raw overlap_auc
# for ea. target, choose source w. higher mean
```

```python
fn = 'run_2c/analysis/results.csv'
df = pd.read_csv(fn)
df['pair'] = df['source'] + '_' + df['target']
df.head()
```

```python
raw_mean_df = df[['pair', 'raw_overlap']].groupby('pair').mean().reset_index()
raw_mean_df = raw_mean_df.rename(columns={'raw_overlap': 'mean_raw_overlap'})
raw_mean_df
```

```python
df = df.merge(raw_mean_df, on='pair')
df.head()
```
