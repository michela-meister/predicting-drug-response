```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import pickle
```

```python
def r_squared(means, test):
    pearson_corr = np.corrcoef(test, means)
    r = pearson_corr[0, 1]
    return np.power(r, 2)

def read_pickle(fn):
    with open(fn, 'rb') as handle:
        val = pickle.load(handle)
    return val
```

```python
df = pd.read_csv('../data/rep-gdsc-ctd2-mean.csv')
df.head()
```

```python
baseline = r_squared(df['REP_published_auc_mean'].to_numpy(), df['GDSC_published_auc_mean'].to_numpy())
print(baseline)
```

```python
train_arr = np.loadtxt('../results/2023-07-20/transfer_multi_eval/REP_GDSC/train_rsq_arr.txt')
test_arr = np.loadtxt('../results/2023-07-20/transfer_multi_eval/REP_GDSC/test_rsq_arr.txt')
```

```python
x_axis = range(1, len(train_arr) + 1)
plt.plot(x_axis, train_arr, 'r-o', label='train')
plt.plot(x_axis, test_arr, 'b-+', label='test')
plt.plot(x_axis, baseline * np.ones(len(x_axis)), 'c-', label='baseline')
plt.ylabel('squared pearson correlation')
plt.xlabel('dimension of transfer matrix')
plt.legend(loc="lower right")
title = 'squared pearson correlation vs dimension of transfer matrix\n REP_published_auc --> GDSC_published_auc \n10 runs per data point'
plt.title(title)
```

```python
# print train, test for each seed
data_directory = '../results/2023-07-20/transfer_multi_eval/REP_GDSC' # then /r/1/seed
# for each r, get out values for all the seeds
```

```python
def results_for_r(directory, r, n_seeds):
    rsq_train = []
    rsq_test = []
    for seed in range(1, n_seeds + 1):
        fn = directory + '/' + str(r) + '/1/' + str(seed) + '.pkl'
        data_dict = read_pickle(fn)
        rsq_train.append(data_dict['rsq_train'])
        rsq_test.append(data_dict['rsq_test'])
    return rsq_train, rsq_test
```

```python
train1, test1 = results_for_r(data_directory, 1, 10)
train2, test2 = results_for_r(data_directory, 2, 10)
train3, test3 = results_for_r(data_directory, 3, 10)
train4, test4 = results_for_r(data_directory, 4, 10)
train5, test5 = results_for_r(data_directory, 5, 10)
train6, test6 = results_for_r(data_directory, 6, 10)
train7, test7 = results_for_r(data_directory, 7, 10)
train8, test8 = results_for_r(data_directory, 8, 10)
```

```python
plt.clf()
x_axis = range(1, 11)
plt.plot(x_axis, train1, 'r-o', label='r=1')
plt.plot(x_axis, test1, 'r-+', label='r=1')
plt.plot(x_axis, train2, 'g-o', label='r=2')
plt.plot(x_axis, test2, 'g-+', label='r=2')
plt.plot(x_axis, train3, 'c-o', label='r=3')
plt.plot(x_axis, test3, 'c-+', label='r=3')
plt.plot(x_axis, train4, 'k-o', label='r=4')
plt.plot(x_axis, test4, 'k-+', label='r=4')
plt.plot(x_axis, train5, 'b-o', label='r=5')
plt.plot(x_axis, test5, 'b-+', label='r=5')
plt.plot(x_axis, train6, 'm-o', label='r=6')
plt.plot(x_axis, test6, 'm-+', label='r=6')
plt.plot(x_axis, train7, 'y-o', label='r=7')
plt.plot(x_axis, test7, 'y-+', label='r=7')
plt.plot(x_axis, train8, 'k', label='r=8')
plt.plot(x_axis, test8, 'k', label='r=8')
plt.xlabel('seed')
plt.ylabel('squared pearson correlation')
plt.legend(loc='lower right')
plt.show()
```
