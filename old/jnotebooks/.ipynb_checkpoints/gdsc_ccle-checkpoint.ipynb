```python
ccle_drugs = ccle['Compound'].unique()
gdsc_drugs = gdsc['DRUG_NAME'].unique()
print('num ccle drugs: ' + str(len(ccle_drugs)))
print('num gdsc drugs: ' + str(len(gdsc_drugs)))
```

```python
drug_intersection = set(ccle_drugs).intersection(gdsc_drugs)
print(drug_intersection)
```

```python
ccle_remaining = set(ccle_drugs).difference(drug_intersection)
print(sorted(list(ccle_remaining)))
```
