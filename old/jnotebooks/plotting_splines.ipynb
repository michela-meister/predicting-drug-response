```python
import matplotlib.pyplot as plt
import numpy as np
import pickle
from scipy.interpolate import splrep, BSpline
```

```python
from scipy.interpolate import UnivariateSpline
```

```python
def read_pickle(fn):
    with open(fn, 'rb') as f:
        v = pickle.load(f)
    return v
```

```python
train_fn = '../results/2023-08-07/opt_k_target_only/log_REP/train_avg.pkl'
test_fn = '../results/2023-08-07/opt_k_target_only/log_REP/test_avg.pkl'
```

```python
train = read_pickle(train_fn)
test = read_pickle(test_fn)
```

```python
print(np.argmax(test))
```

```python
y = test
x = range(0, len(y))
plt.clf()
plt.plot(x, y)
plt.plot(np.argmax(y), y[np.argmax(y)], 'co')
argmaxK = np.argmax(y) + 1 # because k is 1-indexed
print(argmaxK)
```

```python
spl = UnivariateSpline(x, y)
newy = spl(x)
plt.plot(x, newy, 'g', lw=3)
plt.plot(np.argmax(newy), newy[np.argmax(newy)], 'co')
spl_argmaxK = np.argmax(newy) + 1 # because k is 1-indexed
print(spl_argmaxK)
```
