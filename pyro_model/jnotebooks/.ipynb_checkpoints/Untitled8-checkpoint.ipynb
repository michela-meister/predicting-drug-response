```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

def low_rank(M, r):
    U, S, Vh = np.linalg.svd(M, full_matrices=True)
    return np.dot(U[:, :r] * S[:r], Vh[:r,:])

def get_singular_values(M):
    U, S, Vh = np.linalg.svd(M, full_matrices=True)
    return S
```

```python
n_samples = 318
n_drugs = 84
rank = 5
# create a random matrix
arr = np.random.randn(n_samples, n_drugs)
orig_sing_vals = get_singular_values(arr)
plt.ylabel('singular value')
plt.xlabel('rank order')
plt.title('singular value vs. rank order')
plt.scatter(range(len(orig_sing_vals)), orig_sing_vals)
```

```python
# convert to a rank-5 matrix
low_arr = low_rank(arr, rank)
sing_vals = get_singular_values(low_arr)
plt.ylabel('singular value')
plt.xlabel('rank order')
plt.title('singular value vs. rank order')
plt.scatter(range(len(sing_vals)), sing_vals)
```

```python
s_idx, d_idx = np.nonzero(low_arr)
# convert to a dataframe
values = low_arr[s_idx, d_idx]
df = pd.DataFrame(data={'sample_id': s_idx, 'drug_id': d_idx, 'value': values})
df.to_csv('~/Documents/research/tansey/msk_intern/pyro_model/data/rank5.csv')
```
