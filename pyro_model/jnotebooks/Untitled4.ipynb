```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import interpolate

x = np.arange(0, 30)
y = np.exp(-x/3.0)
f = interpolate.interp1d(x, y)
```

```python
# for each mid, get x (days) and y (volume fcn)
# convert x, y to numpy arrays
# ensure that days starts with 1 and ends with something > 21
# get f via interpolation
# end_vol = f(21)
# return end_vol
```

```python
xnew = np.arange(0, 9, 0.1)
ynew = f(xnew)   # use interpolation function returned by `interp1d`
plt.plot(xnew, ynew, 'o')
plt.show()
```
